### Formatação dos Dados:

* Definindo Regras de Formatação:

% == 0 = 0, % > 0 = 1;

CI < 3 = 0, CI >= 3 = 1;

IGC < 3 = 0, IGC >= 3 = 1;

variacao mat <= 0 = 0, variacao mat > 0 = 1;

% FIES == 0 = 0, % FIES > 0 = 1;

CURSOS == 0 = 0, CURSOS > 0 = 1;

saldo <= 0 = 0, saldo > 0 = 1;

* Formatando dataset:

```python
df_formated = df[['Natureza', 'Situacao', '%', 'CI N', 'CI V', 'IGC_N', 'IGC_M', 'IGC_V', 'EAD_17', 'variacao mat', '% FIES', 'CURSOS', 'saldo 2017']]

df_formated["%"] = np.where(df_formated["%"] == 0, 0, 1)

df_formated["CI N"] = np.where(df_formated["CI N"] < 3, 0, 1)
df_formated["CI V"] = np.where(df_formated["CI V"] < 3, 0, 1)

df_formated["IGC_N"] = np.where(df_formated["IGC_N"] < 3, 0, 1)
df_formated["IGC_M"] = np.where(df_formated["IGC_M"] < 3, 0, 1)
df_formated["IGC_V"] = np.where(df_formated["IGC_V"] < 3, 0, 1)

df_formated["variacao mat"] = np.where(df_formated["variacao mat"] <= 0, 0, 1)
df_formated["% FIES"] = np.where(df_formated["% FIES"] == 0, 0, 1)
df_formated["CURSOS"] = np.where(df_formated["CURSOS"] == 0, 0, 1)
df_formated["saldo 2017"] = np.where(df_formated["saldo 2017"] <= 0, 0, 1)

df_formated
```
