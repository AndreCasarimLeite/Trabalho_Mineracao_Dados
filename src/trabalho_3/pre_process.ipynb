# Re-processamento da Base de Dados

### Importação das Bibliotecas:

```python
import pandas as pd
import numpy as np
import seaborn as sns
from pandas import factorize
import re
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

-----------------------------

### Importação dos Datasets:

```python
df = pd.DataFrame()
df = pd.read_csv('../../data/trabalho2_dados10.csv')
df
```

---------------------------

### Tratando Dados Categóricos:

```python
df['Situacao'].loc[df['Situacao'] == 'Ativa'] = 1
df['Situacao'].loc[df['Situacao'] == 'Extinta'] = 0
df['Situacao'] = df['Situacao'].astype('int')
df
```

------------------------------

### Tratando Dados Faltantes:

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df.replace('S/D', np.nan, inplace=True)
df.replace('S/C', np.nan, inplace=True)
```

```python
df_yes_nan = df.loc[df.isna().any(axis=1)]
df_not_nan = df.dropna(axis=0)

target_list = ['CI N', 'CI V', 'IGC_N', 'IGC_M', 'IGC_V']
independent_list = ['Natureza', 'Situacao', 'DIAS', '%', 'EAD_17', 'variacao mat', 'mat_T 2017',
                    '% FIES', 'CURSOS', 'saldo 2017', 'Variacao 17-16']
error_list = []
score_list = []

for i in range(5):
    target_col = target_list[i]
    X_train, X_test, y_train, y_test = train_test_split(df_not_nan[independent_list], df_not_nan[target_col], test_size=0.2, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_input = model.predict(df_yes_nan[independent_list])
    df_yes_nan[target_col] = y_input

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    error_list.append(mse)
    score_list.append(r2)

    if (i < 2):
        df_yes_nan[target_col] = df_yes_nan[target_col].astype(int)
    
```

```python
df = pd.concat([df_yes_nan, df_not_nan])
df
```

--------------------------------

### Tratamento de Dados Inconsistentes:

* Tranformando tipos dos atributos:

```python
df.dtypes
```

```python
target_list = ['CI N', 'CI V', 'IGC_N', 'IGC_M', 'IGC_V']

for i in range(5):
    if i < 2:
        df[target_list[i]] = df[target_list[i]].astype('int64')
    else:
        df[target_list[i]] = df[target_list[i]].astype('float64')
```

* Tratando dados menores que zero:

```python
counts = (df < 0).sum(axis=0)
counts
```

```python
target_list = ['IGC_N', 'IGC_M', 'CURSOS']
for i in range(3):
    df[target_list[i]].loc[df[target_list[i]] < 0] = df[target_list[i]].abs()
```

* Tratando notas maiores que 5:

```python
counts = (df > 5).sum(axis=0)
counts
```

```python
target_list = ['CI N', 'CI V', 'IGC_N', 'IGC_M', 'IGC_V']
for i in range(5):
    df[target_list[i]].loc[df[target_list[i]] > 5] = 5
```

* Tratando notas == 0:

```python
counts = (df == 0).sum(axis=0)
counts
```

```python
target_list = ['CI N', 'CI V', 'IGC_N', 'IGC_M', 'IGC_V']
for i in range(5):
    df[target_list[i]].loc[df[target_list[i]] == 0] = 1
```

-------------------------

### Normalizando os Dados:

```python
df
```

```python
target_list = ['DIAS', '%',
               'mat_T 2017', '% FIES', 'CURSOS']

scaler = MinMaxScaler(feature_range=(0,1))

df_scaled_1 = scaler.fit_transform(df[target_list])
df_scaled_1 = pd.DataFrame(df_scaled_1, columns = [target_list])

df[target_list] = df_scaled_1[target_list]

df
```
